```python
df = df.sort_values("wilson_lower_bound", ascending = False)
df.drop(columns='Unnamed: 0' , inplace = True)
df.head()
```

```python
def missing_values_analysis(df):
    na_columns_ = [col for col in df.columns if df[col].isnull().sum() > 0]  # Corrected variable name
    n_miss = df[na_columns_].isnull().sum().sort_values(ascending=True)  # Corrected variable name
    ratio_ = (df[na_columns_].isnull().sum() / df.shape[0] * 100).sort_values(ascending=True)  # Corrected variable name
    missing_df = pd.concat([n_miss, np.round(ratio_, 2)], axis=1, keys=["Missing Values", "Ratio"])
    return missing_df

def check_dataframe(df):
    print("SHAPE".center(82, '~'))
    print('Rows:', df.shape[0])  # Corrected formatting
    print('Columns:', df.shape[1])  # Corrected formatting
    print("TYPES".center(82, '~'))
    print(df.dtypes)
    print("".center(82, '~'))
    print(missing_values_analysis(df))  # Assuming missing_values_analysis is a defined function
    print('DUPLICATED VALUES'.center(83, '~'))
    print(df.duplicated().sum())
    print("QUANTILES".center(82, '~'))
    print(df.quantile([0, 0.05, 0.50, 0.95, 0.99, 1]))

# Assuming 'df' is your DataFrame
check_dataframe(df)
```

```python
def check_class(dataframe):
    nunique_df = pd.DataFrame({'Variable': dataframe.columns,
                               'Classes': [dataframe[i].nunique() for i in dataframe.columns]})
    nunique_df = nunique_df.sort_values('Classes', ascending=False)
    nunique_df = nunique_df.reset_index(drop=True)
    return nunique_df

# Assuming 'df' is your DataFrame
result = check_class(df)
print(result)
```

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

def categorical_variable_summary(df, column_name):
    fig = make_subplots(rows=1, cols=2,
                        subplot_titles=('Countplot', 'Percentage'),
                        specs=[[{"type": "xy"}, {"type": "domain"}]])
    fig.add_trace(go.Bar(y=df[column_name].value_counts().values.tolist(),
                         x=[str(i) for i in df[column_name].value_counts().index],
                         text=df[column_name].value_counts().values.tolist(),
                         marker=dict(color='brown',
                                     line=dict(color='black', width=1))),
                  row=1, col=1)
    fig.add_trace(go.Pie(labels=df[column_name].value_counts().keys(),
                          values=df[column_name].value_counts().values,
                          textfont=dict(size=18),
                          marker=dict(colors=px.colors.qualitative.Plotly)),
                  row=1, col=2)

    fig.update_layout(title_text='Categorical Variable Analysis',
                      showlegend=False)
    fig.show()

categorical_variable_summary(df,'overall')
```

```python
rt = lambda x: re.sub("[^a-zA-Z]", '' ,str(x))
df["reviewText"] = df["reviewText"].map(rt)
df["reviewText"] = df["reviewText"].str.lower()
df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Assuming df is your DataFrame and "reviewText" is the column containing text reviews

analyzer = SentimentIntensityAnalyzer()

# Iterate over each review text in the "reviewText" column
for index, row in df["reviewText"].items():
    # Get sentiment scores for the current review text
    score = analyzer.polarity_scores(row)
    
    # Extract scores for negative, neutral, and positive sentiments
    neg = score["neg"]
    neu = score["neu"]
    pos = score["pos"]
    
    # Determine sentiment based on scores
    if neg > pos:
        df.loc[index, "sentiment"] = "Negative"
    elif pos > neg:
        df.loc[index, "sentiment"] = "Positive"
    else:
        df.loc[index, "sentiment"] = "Neutral"
```
